**Task - 3 Outlier Detection**

Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
import pandas as pd
import matplotlib.pyplot as plt


south_asian_data = pd.read_csv('south_asian_countries.csv')



Q1_score = south_asian_data['score'].quantile(0.25)
Q3_score = south_asian_data['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = south_asian_data['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asian_data['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp


lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp


outliers_score = south_asian_data[(south_asian_data['score'] < lower_bound_score) |
                                  (south_asian_data['score'] > upper_bound_score)]

outliers_gdp = south_asian_data[(south_asian_data['Log GDP per capita'] < lower_bound_gdp) |
                                (south_asian_data['Log GDP per capita'] > upper_bound_gdp)]

outliers = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()

plt.figure(figsize=(10, 6))


plt.scatter(south_asian_data['Log GDP per capita'], south_asian_data['score'], label='Normal', color='blue')


plt.scatter(outliers['Log GDP per capita'], outliers['score'], label='Outliers', color='red', s=100, marker='X')


plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.title('GDP per Capita vs Score with Outliers Highlighted')
plt.legend()
plt.show()


print("Outliers based on Score and GDP per Capita:")
print(outliers[['Country name', 'score', 'Log GDP per capita']])
```

**Task - 4 - Exploring Trends Across Metrics:**

Tasks:
1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation
{pearson correlation} with the Score for South Asian countries.
2. Create scatter plots with trendlines for these metrics against the Score.
3. Identify and discuss the strongest and weakest relationships between these metrics and the Score for
South Asian countries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression


south_asian_data = pd.read_csv('south_asian_countries.csv')


metric_1 = 'Freedom to make life choices'
metric_2 = 'Generosity'


correlation_1 = pearsonr(south_asian_data[metric_1], south_asian_data['score'])[0]
correlation_2 = pearsonr(south_asian_data[metric_2], south_asian_data['score'])[0]

print(f"Pearson correlation between {metric_1} and Score: {correlation_1:.2f}")
print(f"Pearson correlation between {metric_2} and Score: {correlation_2:.2f}")


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
sns.scatterplot(data=south_asian_data, x=metric_1, y='score', color='blue')
sns.regplot(data=south_asian_data, x=metric_1, y='score', scatter=False, color='red', line_kws={'lw': 2})
plt.title(f"{metric_1} vs Score with Trendline")
plt.xlabel(metric_1)
plt.ylabel("Score")

sns.scatterplot(data=south_asian_data, x=metric_2, y='score', color='green')
sns.regplot(data=south_asian_data, x=metric_2, y='score', scatter=False, color='orange', line_kws={'lw': 2})
plt.title(f"{metric_2} vs Score with Trendline")
plt.xlabel(metric_2)
plt.ylabel("Score")

plt.tight_layout()
plt.show()


if abs(correlation_1) > abs(correlation_2):
    print(f"The strongest relationship is between {metric_1} and Score with a correlation of {correlation_1:.2f}")
else:
    print(f"The strongest relationship is between {metric_2} and Score with a correlation of {correlation_2:.2f}")

if abs(correlation_1) < abs(correlation_2):
    print(f"The weakest relationship is between {metric_1} and Score with a correlation of {correlation_1:.2f}")
else:
    print(f"The weakest relationship is between {metric_2} and Score with a correlation of {correlation_2:.2f}")
```

2. Top and Bottom Performers:
• Identify the top 3 and bottom 3 countries in each region based on the score.
• Plot bar charts comparing these charts.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/drive/MyDrive/AI /WHR-2024-5CS037.csv'
data = pd.read_csv(file_path)

# Check the column names to ensure proper filtering
print("Column names in the dataset:", data.columns)

# Define South Asian and Middle Eastern countries
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                         "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

middle_eastern_countries = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan", "Kuwait",
                            "Lebanon", "Oman", "Palestine", "Qatar", "Saudi Arabia",
                            "Syria", "United Arab Emirates", "Yemen"]

# Assign the Region column based on country names
data['Region'] = data['Country name'].apply(
    lambda x: 'South Asia' if x in south_asian_countries else ('Middle East' if x in middle_eastern_countries else 'Other')
)

# Group by Region and identify top 3 and bottom 3 countries in each region
regions = data['Region'].unique()
top_bottom_performers = {}

for region in regions:
    if region == 'Other':  # Skip regions not relevant to South Asia or Middle East
        continue

    region_data = data[data['Region'] == region]

    # Sort by score and get the top 3 and bottom 3 countries
    top_3 = region_data.nlargest(3, 'score')
    bottom_3 = region_data.nsmallest(3, 'score')

    top_bottom_performers[region] = {
        'Top 3': top_3[['Country name', 'score']],
        'Bottom 3': bottom_3[['Country name', 'score']]
    }

# Plot Bar Charts for each region
for region, performers in top_bottom_performers.items():
    fig, ax = plt.subplots(figsize=(10, 6))

    # Combine top 3 and bottom 3 into one DataFrame for easy plotting
    combined = pd.concat([performers['Top 3'], performers['Bottom 3']])

    # Plot bar chart
    combined.plot(kind='bar', x='Country name', y='score', ax=ax, color=['green']*3 + ['red']*3, legend=False)

    ax.set_title(f"Top 3 and Bottom 3 Countries in {region}", fontsize=16)
    ax.set_xlabel('Country', fontsize=12)
    ax.set_ylabel('Score', fontsize=12)
    ax.set_xticklabels(combined['Country name'], rotation=45, ha='right')
    ax.grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout()
    plt.show()

```

6. Outlier Detection:
• Identify outlier countries in both regions based on Score and GDP per Capita.
• Plot these outliers and discuss their implications.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace with the actual file path)
file_path = '/content/drive/MyDrive/AI /WHR-2024-5CS037.csv'
data = pd.read_csv(file_path)

# Define South Asian and Middle Eastern countries
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                         "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

middle_eastern_countries = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan", "Kuwait",
                            "Lebanon", "Oman", "Palestine", "Qatar", "Saudi Arabia",
                            "Syria", "United Arab Emirates", "Yemen"]

# Assign the Region column based on country names
data['Region'] = data['Country name'].apply(
    lambda x: 'South Asia' if x in south_asian_countries else ('Middle East' if x in middle_eastern_countries else 'Other')
)

# Filter the data to focus on South Asia and Middle East regions
regions_data = data[data['Region'].isin(['South Asia', 'Middle East'])]

# Define a function to detect outliers using IQR method
def detect_outliers(df, column):
    # Calculate Q1, Q3, and IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    return outliers

# Detect outliers for 'score' and 'Log GDP per Capita' in both regions
score_outliers = detect_outliers(regions_data, 'score')
gdp_outliers = detect_outliers(regions_data, 'Log GDP per capita')

# Plot the outliers on a scatter plot: Score vs. Log GDP per Capita
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
ax.scatter(regions_data['Log GDP per capita'], regions_data['score'], color='blue', alpha=0.7, label='Data')

# Plot the outliers
ax.scatter(score_outliers['Log GDP per capita'], score_outliers['score'], color='red', label='Score Outliers', s=100)
ax.scatter(gdp_outliers['Log GDP per capita'], gdp_outliers['score'], color='green', label='GDP Outliers', s=100, marker='*')

# Add labels and title
ax.set_xlabel('Log GDP per Capita', fontsize=12)
ax.set_ylabel('Score', fontsize=12)
ax.set_title('Outliers in South Asia and Middle East based on Score and GDP per Capita', fontsize=14)
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()

# Display the outliers
print("Score Outliers:")
print(score_outliers[['Country name', 'score', 'Log GDP per capita']])

print("\nGDP Outliers:")
print(gdp_outliers[['Country name', 'score', 'Log GDP per capita']])
```
