**Task - 3 Outlier Detection**

Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5  IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
import pandas as pd
import matplotlib.pyplot as plt


south_asian_data = pd.read_csv('south_asian_countries.csv')



Q1_score = south_asian_data['score'].quantile(0.25)
Q3_score = south_asian_data['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = south_asian_data['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asian_data['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp


lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp


outliers_score = south_asian_data[(south_asian_data['score'] < lower_bound_score) |
                                  (south_asian_data['score'] > upper_bound_score)]

outliers_gdp = south_asian_data[(south_asian_data['Log GDP per capita'] < lower_bound_gdp) |
                                (south_asian_data['Log GDP per capita'] > upper_bound_gdp)]

outliers = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()

plt.figure(figsize=(10, 6))


plt.scatter(south_asian_data['Log GDP per capita'], south_asian_data['score'], label='Normal', color='blue')


plt.scatter(outliers['Log GDP per capita'], outliers['score'], label='Outliers', color='red', s=100, marker='X')


plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.title('GDP per Capita vs Score with Outliers Highlighted')
plt.legend()
plt.show()


print("Outliers based on Score and GDP per Capita:")
print(outliers[['Country name', 'score', 'Log GDP per capita']])
```

**Task - 4 - Exploring Trends Across Metrics:**

Tasks:
1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation
{pearson correlation} with the Score for South Asian countries.
2. Create scatter plots with trendlines for these metrics against the Score.
3. Identify and discuss the strongest and weakest relationships between these metrics and the Score for
South Asian countries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression


south_asian_data = pd.read_csv('south_asian_countries.csv')


metric_1 = 'Freedom to make life choices'
metric_2 = 'Generosity'


correlation_1 = pearsonr(south_asian_data[metric_1], south_asian_data['score'])[0]
correlation_2 = pearsonr(south_asian_data[metric_2], south_asian_data['score'])[0]

print(f"Pearson correlation between {metric_1} and Score: {correlation_1:.2f}")
print(f"Pearson correlation between {metric_2} and Score: {correlation_2:.2f}")


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
sns.scatterplot(data=south_asian_data, x=metric_1, y='score', color='blue')
sns.regplot(data=south_asian_data, x=metric_1, y='score', scatter=False, color='red', line_kws={'lw': 2})
plt.title(f"{metric_1} vs Score with Trendline")
plt.xlabel(metric_1)
plt.ylabel("Score")

sns.scatterplot(data=south_asian_data, x=metric_2, y='score', color='green')
sns.regplot(data=south_asian_data, x=metric_2, y='score', scatter=False, color='orange', line_kws={'lw': 2})
plt.title(f"{metric_2} vs Score with Trendline")
plt.xlabel(metric_2)
plt.ylabel("Score")

plt.tight_layout()
plt.show()


if abs(correlation_1) > abs(correlation_2):
    print(f"The strongest relationship is between {metric_1} and Score with a correlation of {correlation_1:.2f}")
else:
    print(f"The strongest relationship is between {metric_2} and Score with a correlation of {correlation_2:.2f}")

if abs(correlation_1) < abs(correlation_2):
    print(f"The weakest relationship is between {metric_1} and Score with a correlation of {correlation_1:.2f}")
else:
    print(f"The weakest relationship is between {metric_2} and Score with a correlation of {correlation_2:.2f}")
```

6. Outlier Detection:
 Identify outlier countries in both regions based on Score and GDP per Capita.
 Plot these outliers and discuss their implications.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


south_asian_data = pd.read_csv('south_asian_countries.csv')
middle_east_data = pd.read_csv('middle_east_countries.csv')


def identify_outliers(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]
    return outliers


outliers_south_asia_score = identify_outliers(south_asian_data, 'score')
outliers_south_asia_gdp = identify_outliers(south_asian_data, 'Log GDP per capita')

outliers_middle_east_score = identify_outliers(middle_east_data, 'score')
outliers_middle_east_gdp = identify_outliers(middle_east_data, 'Log GDP per capita')

plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.scatter(south_asian_data['Log GDP per capita'], south_asian_data['score'], label='Data', color='blue')
plt.scatter(outliers_south_asia_gdp['Log GDP per capita'], outliers_south_asia_score['score'],
            color='red', label='Outliers')
plt.title('South Asia: Score vs GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(middle_east_data['Log GDP per capita'], middle_east_data['score'], label='Data', color='blue')
plt.scatter(outliers_middle_east_gdp['Log GDP per capita'], outliers_middle_east_score['score'],
            color='red', label='Outliers')
plt.title('Middle East: Score vs GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.legend()

plt.tight_layout()
plt.show()


print("South Asia Outliers (Score and GDP per Capita):")
print(outliers_south_asia_score[['Country name', 'score', 'Log GDP per capita']])

print("\nMiddle East Outliers (Score and GDP per Capita):")
print(outliers_middle_east_score[['Country name', 'score', 'Log GDP per capita']])
```
